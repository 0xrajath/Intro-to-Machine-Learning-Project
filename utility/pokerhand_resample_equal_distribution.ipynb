```python
df_fok = pd.read_csv('../data/generated_hand_data/four_of_a_kind.csv', header=None)
df_fok = df_fok.sample(frac=0.017).reset_index(drop=True)
df_fok
```

```python
# Python function to print permutations of a given list 
def permutation(lst): 
  
    # If lst is empty then there are no permutations 
    if len(lst) == 0: 
        return [] 
  
    # If there is only one element in lst then, only 
    # one permuatation is possible 
    if len(lst) == 1: 
        return [lst] 
  
    # Find the permutations for lst if there are 
    # more than 1 characters 
  
    l = [] # empty list that will store current permutation 
  
    # Iterate the input(lst) and calculate the permutation 
    for i in range(len(lst)): 
       m = lst[i] 
  
       # Extract lst[i] or m from the list.  remLst is 
       # remaining list 
       remLst = lst[:i] + lst[i+1:] 
  
       # Generating all permutations where m is first 
       # element 
       for p in permutation(remLst): 
           l.append([m] + p) 
    return l 
```

```python
# Permute a given hand

def permute_hand(df,hand):
    temp=[]

    for row in df.iterrows():
        index, data = row
        temp.append(data.tolist())

    tuple_list = []

    for rowList in temp:
        tuple_list.append([(rowList[0],rowList[1]),(rowList[2],rowList[3]),(rowList[4],rowList[5]),(rowList[6],rowList[7]),(rowList[8],rowList[9])])

    #print(len(tuple_list))
    final_list = []

    for rowList in tuple_list:
        perm_list = permutation(rowList)
        for perm in perm_list:
            final_list_row = []
            final_list_row.append(perm[0][0])
            final_list_row.append(perm[0][1])
            final_list_row.append(perm[1][0])
            final_list_row.append(perm[1][1])
            final_list_row.append(perm[2][0])
            final_list_row.append(perm[2][1])
            final_list_row.append(perm[3][0])
            final_list_row.append(perm[3][1])
            final_list_row.append(perm[4][0])
            final_list_row.append(perm[4][1])
            final_list_row.append(hand)
            final_list.append(final_list_row)


    return final_list

```

```python
# New Full House Dataframe
fh_perm = permute_hand(df_fh,6)

df_fh_permute =pd.DataFrame(fh_perm,columns=["S1","C1","S2","C2","S3","C3","S4","C4","S5","C5","hand"])
df_fh_permute = df_fh_permute.sample(frac=0.28).reset_index(drop=True)

df_fh_new = pd.concat([df_fh, df_fh_permute])
df_fh_new = df_fh_new.reset_index(drop=True)
df_fh_new.drop_duplicates(keep=False, inplace=True)
df_fh_new = df_fh_new.reset_index(drop=True)
df_fh_new
```

```python
# New Flush Dataframe
f_perm = permute_hand(df_f,5)

df_f_permute =pd.DataFrame(f_perm,columns=["S1","C1","S2","C2","S3","C3","S4","C4","S5","C5","hand"])
df_f_permute = df_f_permute.sample(frac=0.19).reset_index(drop=True)

df_f_new = pd.concat([df_f, df_f_permute])
df_f_new = df_f_new.reset_index(drop=True)
df_f_new.drop_duplicates(keep=False, inplace=True)
df_f_new = df_f_new.reset_index(drop=True)
df_f_new
```

```python
# New Straight Dataframe
s_perm = permute_hand(df_s,4)

df_s_permute =pd.DataFrame(s_perm,columns=["S1","C1","S2","C2","S3","C3","S4","C4","S5","C5","hand"])
df_s_permute = df_s_permute.sample(frac=0.11).reset_index(drop=True)

df_s_new = pd.concat([df_s, df_s_permute])
df_s_new = df_s_new.reset_index(drop=True)
df_s_new.drop_duplicates(keep=False, inplace=True)
df_s_new = df_s_new.reset_index(drop=True)
df_s_new
```

```python
# New Three of a Kind Dataframe
tok_perm = permute_hand(df_tok,3)

df_tok_permute =pd.DataFrame(tok_perm,columns=["S1","C1","S2","C2","S3","C3","S4","C4","S5","C5","hand"])
df_tok_permute = df_tok_permute.sample(frac=0.012).reset_index(drop=True)

df_tok_new = pd.concat([df_tok, df_tok_permute])
df_tok_new = df_tok_new.reset_index(drop=True)
df_tok_new.drop_duplicates(keep=False, inplace=True)
df_tok_new = df_tok_new.reset_index(drop=True)
df_tok_new
```

```python
df_train_final = pd.concat([df_987, df_nothing, df_op, df_tp, df_fh_new, df_f_new, df_s_new, df_tok_new])
df_train_final = df_train_final.sample(frac=1).reset_index(drop=True)
df_train_final
```
